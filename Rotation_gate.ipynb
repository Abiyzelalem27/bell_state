Arbitrary axis rotation

rotate around any axis [nx, ny, nz]. For example, along (1,1,1)/√3:

```python
theta = np.pi / 2  # 90 degrees
axis = [1/np.sqrt(3), 1/np.sqrt(3), 1/np.sqrt(3)]
U= rotation_gate(theta, axis)
print("U:\n", U)
```

Composite rotations
You can combine rotations to form more complex gates. Example: Rx followed by R

```python
theta= np.pi / 2
Rx = rotation_gate(theta, [1, 0, 0])
Ry = rotation_gate(theta, [0, 1, 0])
U = np.dot(Ry, Rx)  
print("U:\n", U)
```
