Universal Single-Qubit Gate Decomposition

The code in this notebook demonstrates how to synthesize the standard Pauli gates (X, Y, Z) and the Phase gate (S) using only Hadamard (H) and T gates. These are the basic building blocks of universal quantum computing.
Note: Small numerical values like 1.732e-16 are floating-point errors and should be treated as 0.

```python
I= H @ H
print("I:\n", I)
```

```python
I= T @ T
print("I:\n", I)
```

```python
# X using H and T only
X= H @ T @ T  @ T @ T @ H
print("X:\n", X)
```

```python
# Z using T only
Z= T @ T  @ T @ T 
print("Z:\n", Z)
```

```python
Y = -1j * H @ T @ T @ T @ T @ H
print("Y:\n", Y)
```

```python
S = T @ T
print("S:\n", S)
```

# Universal Gate Set for Quantum Computing
To build any quantum computer, a small set of gates is sufficient:

## 1. Single-Qubit Gates: H + T 
- The H and T (π/8) gates can generate any single-qubit gate.  
- By combining them in sequences, you can synthesize X, Y, Z, S, and any rotation.

## 2. Entangling Gate: CNOT
- The CNOT gate introduces entanglement between qubits.  
- Combined with H and T, it makes the set universal for multi-qubit circuits.

## 3. Floating-Point Artifacts
- In simulations, very small numerical errors like `1.7e-16` may appear.  
- These are negligible and do not affect logical correctness.

Note: Using only H, T, and CNOT, you can construct any quantum circuit, demonstrating the power of gate decomposition and the universality of this minimal gate set.
